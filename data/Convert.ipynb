# all entity coreference anaphors

```python
import collections
ar_anaphors = collections.Counter()
cnt = 0
max_len = 0
for doc in doc_dict:
    visited = {}
    coref_stack = []
    for line_id, line in enumerate(doc_dict[doc]):
        if line[-5] != '_':
            parts = line[-5].split('(')
            if len(parts[0]):
                for _ in range(len(parts[0])):
                    entity_id, st = coref_stack.pop()
                    ed = line_id
                    if (entity_id not in visited) or ('-Pseudo' in entity_id):
                        visited[entity_id] = True
                        continue
                    word = []
                    for i in range(st, ed+1):
                        word += [doc_dict[doc][i][1]]
                    max_len = max(max_len, len(word))
                    word = ' '.join(word)
                    ar_anaphors[word.lower()] += 1
                    cnt += 1
                    
            for part in parts[1:]:
                split_part = part.split('|')
                entity_id = split_part[0]
                st = line_id
                if part[-1] == ')':
                    ed = line_id
                    if (entity_id not in visited) or ('-Pseudo' in entity_id):
                        visited[entity_id] = True
                        continue
                    word = []
                    for i in range(st, ed+1):
                        word += [doc_dict[doc][i][1]]
                    max_len = max(max_len, len(word))
                    word = ' '.join(word)
                    ar_anaphors[word.lower()] += 1
                    cnt += 1
                else:
                    coref_stack.append((entity_id, st))
cnt, max_len, len(ar_anaphors)
```
